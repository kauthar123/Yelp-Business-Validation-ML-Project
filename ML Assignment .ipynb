#  Here our target column is is_open through we decide the business is open or not 

# This is a binary classification problem 

```python
merged_df.head(3)
```

```python
columns_to_drop = ['business_id', 'user_id', 'address','name_x','name_y','postal_code','friends','text','attributes','categories','date_x','date_y','yelping_since']  

# Drop the columns and retain the rest
merged_df = merged_df.drop(columns=columns_to_drop)

```

```python
merged_df.head(3)
```

#  Splitting dataset into x and y

```python
x = merged_df.drop(['is_open'],axis=1)
y = merged_df['is_open']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print("The shape of x_train is",x_train.shape)
print("The shape of x_test is",x_test.shape)
print("The shape of y_train is",y_train.shape)
print("The shape of y_test is",y_test.shape)
```
